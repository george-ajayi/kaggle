<h3> Task #5 Which actor/actress shows up most frequently in the netflix database

```python
df=pd.read_csv('~/Documents/netflix_titles.csv')
casts=df['cast'].reset_index()
casts=casts.cast.str.lower().str.split(',').explode().reset_index()

casts=casts.groupby('cast').agg({'cast':'count'})
casts.rename(columns={'cast':'count'},inplace=True)
casts.reset_index().sort_values('count',ascending=False).head(50)
```
